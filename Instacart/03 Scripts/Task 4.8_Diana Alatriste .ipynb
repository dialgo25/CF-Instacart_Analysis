# 04. 4.8.4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the `transform()` function on the “order_number” column to generate the maximum orders for each user.
# 3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create 1st condition

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Create 2nd condition

ords_prods_merge.loc[(ords_prods_merge['max_order']<= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Create 3rd condition

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency of 'loyalty_flag' newly createe column

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# 06. 4.8.6.

### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the `transform()` function on the “prices” column to find the average price per order for each user.
# 3. Create a new column, “average_expenses,” into which you’ll place the results of your aggregation.

ords_prods_merge['average_expenses'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output

ords_prods_merge.head()
```

```python
# Check frequency of average_expenses column

ords_prods_merge['average_expenses'].value_counts(dropna = False)
```

```python
# check description of 'average_expenses'

ords_prods_merge.describe()
```

```python
# Create 1st condition

ords_prods_merge.loc[ords_prods_merge['average_expenses'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Create 2nd condition

ords_prods_merge.loc[ords_prods_merge['average_expenses'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check output

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 07. 4.8.7

### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the `transform()` function on the “days_since_prior_order” (aka - "last_purchase") column to calculate ordering behaviour for each user (via median of 'last_purchase' column.
# 3. Create a new column, “user_behaviour,” into which you’ll place the results of your aggregation.

ords_prods_merge['user_behaviour'] = ords_prods_merge.groupby(['user_id'])['last_purchase'].transform(np.median)
```

```python
# Check output

ords_prods_merge.head()
```

```python
# Create 1st condition

ords_prods_merge.loc[ords_prods_merge['user_behaviour'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Create 2nd condition

ords_prods_merge.loc[(ords_prods_merge['user_behaviour'] > 10) & (ords_prods_merge['user_behaviour'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Create 3rd condition

ords_prods_merge.loc[ords_prods_merge['user_behaviour'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check output


ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
