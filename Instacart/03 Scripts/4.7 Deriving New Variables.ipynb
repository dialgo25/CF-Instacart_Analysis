# 03. Defining a new user-defined function

```python
#Write the if-statement

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function in dataframe subset

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Counting frequency of values in newly created 'price_range' column

df['price_range'].value_counts(dropna = False)
```

```python
# Apply loop on Instacart dataframe 'ords_prods_merged'
# First we will create a new column that summarizes how busy each day of the week is
#Start by finding out on which day of the week most orders take place

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

### 0 = Saturday; 4 = Wednesday

```python
#Create a new column that will contain one of three different values: “Busiest day,” “Least busy,” and “Regularly busy.”

result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Print result of previous loop

result
```

```python
# Create new column in ords_prods_merged and set it equal to 'result'

ords_prods_merged['busiest_day'] = result
```

```python
# Check frequency of values in newly created column

ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
# Create 3 groups based on frequency of order: Most orders = 10, 11, 14, 15, 13, 12, 16, 9; Average orders: 17, 8, 18, 19, 20, 7, 21, 22; Fewest orders: 23, 6, 0, 1, 5, 2, 4, 3.

hod = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hod.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hod.append("Fewest orders")
  else:
    hod.append("Average orders")
```
